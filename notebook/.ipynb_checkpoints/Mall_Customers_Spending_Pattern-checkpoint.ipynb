### 5: Customer Segmentation
Group customers into three segments based on their Spending Score:
Low (1-33)
Medium (34-66)
High (67-100)
Compute the average Age and Annual Income for each segment. What are the characteristics of each group?
Propose marketing strategies for each segment.
Hint: Use NumPy’s slicing and aggregation techniques.

Instead of finding segment wide avg use for loop based on the segments classifications

```python
# Define the segments
segments = [
    ("Low", 0, 33),
    ("Medium", 34, 66),
    ("High", 67, 100)
]

# Iterate over the segments
for name, min_score, max_score in segments:
    segment = Cust_data[(Cust_data[:, 4] >= min_score) & (Cust_data[:, 4] <= max_score)]
    avg_age = np.mean(segment[:, 2])
    avg_income = np.mean(segment[:, 3])

    print(f"{name} Spending Score Segment:")
    print(f"Average Age: {avg_age:.2f}")
    print(f"Average Annual Income: {avg_income:.2f}\n")
```

### 6: Find Behavioral Similarities
Normalize the dataset (subtract the mean and divide by the standard deviation for each column).
Compute the Euclidean distance between each customer and the customer with the highest Spending Score. Identify customers who exhibit similar behavior.
Hint: Use NumPy’s vectorized operations for distance calculations.

```python
norm_cust_data = Cust_data[:, 2:]
#normalize the age
mean_age = np.mean (norm_cust_data[:, 0])
std_age= np.std (norm_cust_data[:, 0])
norm_age = (norm_cust_data[:, 0]-mean_age)/std_age

#normalize the salary
mean_salary = np.mean (norm_cust_data[:, 1])
std_salary= np.std (norm_cust_data[:, 1])
norm_salary = (norm_cust_data[:, 1]-mean_salary)/std_salary

#normalise the spending score
mean_spend = np.mean (norm_cust_data[:, 2])
std_spend= np.std (norm_cust_data[:, 2])
norm_spend = (norm_cust_data[:, 2]- mean_spend)/std_spend

norm_cust_data = np.array([norm_age, norm_salary, norm_spend]).T
norm_cust_data
```

### Below Code is recommended by meta.ai

```python
# Assuming norm_cust_data is a list or a DataFrame, convert it to a NumPy array
norm_cust_data = np.array(norm_cust_data, dtype=np.float64)

# Find the customer with the highest Spending Score
max_spending_idx = np.argmax(norm_cust_data[:, 2])
max_spending_customer = norm_cust_data[max_spending_idx]

# Compute the Euclidean distance between each customer and the customer with the highest Spending Score
distances = np.linalg.norm(norm_cust_data - max_spending_customer[np.newaxis, :], axis=1)

# Print the customers with the smallest distances (i.e., similar behavior)
similar_customers = Cust_data[np.argsort(distances)[:10]]  # Show the top 10 similar customers
print(similar_customers)
```
